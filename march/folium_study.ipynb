```python
import folium
```

```python
mapa = folium.Map(
    location=[-64.8, -64.1],
    zoom_start=5
)
```

```python
folium.Marker(
    location=[-64.8, -64.1],
    popup="Ilha Biscoe"
).add_to(mapa)
mapa
```

```python
folium.CircleMarker(
location=[-64.8,-64.1],
radius=6,
color="blue",
fill=True,
fill_color="blue",
popup="Biscoe Island"
).add_to(mapa)
mapa
```

```python
import pandas as pd

data = {
"species": ["Adelie","Chinstrap","Gentoo"],
"lat": [-64.8,-64.7,-64.6],
"lon": [-64.1,-64.2,-64.0]
}


df = pd.DataFrame(data)
```

```python
mapa=folium.Map(location=[-64.7,-64.1],zoom_start=6, tiles="CartoDB positron")

for _, row in df.iterrows():
    popup_text=f"""
    Species:{row['species']}
    Latitude:{row['lat']}
    Longitude:{row['lon']}
    """

    folium.CircleMarker(
    location=[row["lat"],row["lon"]],
    radius=6,
    popup=popup_text
    ).add_to(mapa)

mapa

# mapa.save("mapa_penguins.html")
```
