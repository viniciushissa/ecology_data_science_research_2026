```python
import pandas as pd
import plotly.express as px
import seaborn as sns
```

```python
df = pd.read_csv("processed_penguins.csv")
df.head()
```

```python
fig = px.scatter(df, x="flipper_length_mm", y="body_mass_g", color="species",
                 hover_data=["bill_length_mm",
                             "bill_depth_mm",
                             "sex"])
fig.show()
```

```python
fig = px.histogram(df, x="body_mass_g", color="species")
fig.show()
```

```python
fig = px.box(df, x="species", y="body_mass_g", color="island")
fig.show()
```

Exercícios Plotly

```python
fig = px.scatter(df, x="flipper_length_mm", y="body_mass_g", color="species")
fig.update_layout(
    title="Relação entre Flipper e Massa Corporal",
    xaxis_title="Comprimento da Nadadeira (mm)",
    yaxis_title="Massa Corporal (g)"
)
fig.show()
```

```python
fig = px.histogram(df, x="body_mass_g", color="species", facet_col="species")
fig.show()
```

```python
fig = px.violin(df, x="species", y="body_mass_g", color="species")
fig.show()
```

```python
fig = px.scatter(df, x="flipper_length_mm", y="bill_length_mm", color="species",
                 hover_data=["bill_depth_mm", "body_mass_g", "species", "sex"])
fig.show()
```

```python
fig = px.scatter(df,
                 x="flipper_length_mm",
                 y="body_mass_g",
                 color="species",
                 facet_col="species")

fig.show()

# fig.write_html("scatter_penguins.html")
```

Comparação Seaborn vs Plotly
A conclusão é que a biblioteca Seaborn é ideal para fazer apresentação de gráficos para artigos científicos, devido aos seus gráficos serem estáticos e possuir integração com a biblioteca Matplotlib, possibilitando uma edição mais elaborada na montagem da exibição. Já a biblioteca Plotly condiz mais com dashboards, pois possibilita interação com seus gráficos, como exploração mais profunda dos dados da análise em questão e exportação dessas figuras para o formato HTML.
