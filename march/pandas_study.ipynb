```python
def category(x):
    if x < 4000:
        return "light"
    elif x >= 4000 and x < 5000:
        return "medium"
    elif x >= 5000:
        return "heavy"
    else:
        return None

df_tmp4 = df.copy()
df_tmp4["body_mass_category"] = df_tmp4["body_mass_g"].apply(category)
df_tmp4
```

```python
df_tmp6 = df.copy()
df_tmp6.fillna({"sex":"unknown"}, inplace=True)
df_tmp6
```

```python
df_tmp6.fillna({"bill_length_mm": df_tmp6["bill_length_mm"].mean()}, inplace=True)
```

```python
df_tmp6
```

```python
def g_to_kg(x):
    return x/1000

df_tmp6["body_mass_g"] = df_tmp6["body_mass_g"].apply(g_to_kg)
df_tmp6.rename(columns={"body_mass_g":"body_mass_kg"}, inplace=True)
df_tmp6
```

```python
def round_2cases(x):
    return round(x, 2)

df_tmp6 = df_tmp6.apply(round_2cases)
df_tmp6
```

```python
def min_cases(x):
    return x.lower()

df_tmp6["species"] = df_tmp6["species"].apply(min_cases)
df_tmp6
```

```python
df_tmp6.set_index("species", inplace=True)
df_tmp6
```
