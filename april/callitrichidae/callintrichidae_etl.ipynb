Animalia/Chordata/Mammalia/Primates/Callintrichidae

```python
import pandas as pd
```

```python
df = pd.read_csv("callintrichidae_raw.csv", sep="\t")
df.shape
```

```python
df.columns
```

```python
df = df.loc[:, ["genus", "species", "stateProvince", "decimalLatitude", "decimalLongitude", "coordinateUncertaintyInMeters", "eventDate", "month", "year"]]
df.head()
```

```python
df.info()
```

```python
df[df.isna().any(axis=1)]
```

```python
df.fillna(
    {"species": "Unknown",
     "stateProvince": "Unknown",
     "coordinateUncertaintyInMeters": "Unknown"},
     inplace=True
)
df.dropna(inplace=True)
df[df.isna().any(axis=1)]
```

```python
df.value_counts("stateProvince")
```

```python
df["stateProvince"] = df["stateProvince"].replace("Goias", "Goiás")
df["stateProvince"] = df["stateProvince"].replace("Espirito Santo", "Espírito Santo")
df = df[df["stateProvince"] != "Rio Preto da Eva"]
df = df[df["year"] != 2026]
```

```python
df.value_counts("stateProvince")
```

```python
df.to_csv("callintrichidae_clean.csv")
```
